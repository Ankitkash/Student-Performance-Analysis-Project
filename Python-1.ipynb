# Drop/Delete unnamed column

```python
# First column (Unnamed) which is not useful so we will delete it.
df = df.drop("Unnamed: 0", axis = 1)      # axis = 1 -> Vertical axis
print(df.head())
```

# Change weekly study hours column

```python
# We will transform data.
#df["TestPrep"] = df["TestPrep"].str.replace("none", "0")
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10", "5 to 10")
df["TestPrep"] = df["TestPrep"].str.replace("0", "none")
df.head()
```

# Reload csv file

```python
# Reload the data
df = pd.read_csv("students_scores.csv")
df.head(100)
```

```python
# Drop Column Unnamed again
df = df.drop("Unnamed: 0", axis = 1)      # axis = 1 -> Vertical axis
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize = (7, 7))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# From the above chart we have analysed that:
# the number of females in the data is more than the number of males
```

# How Parent Education is affecting students score

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's education and student's score")
plt.show()
```

```python
# From the above chart we concluded that the education of the parents have a good impact on student's scores
```

```python
# Based on the parent marital status
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parent's marital status and students's scores")
plt.show()
```

```python
# From the above chart we concluded that the marital status of the parents have no/negligible impact on student's scores
```

# Distribution of ethnic group

```python
# For % distribution use py chart
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

# Labels
l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

# Test score distributions

```python
sns.histplot(df[['MathScore', 'ReadingScore', 'WritingScore']])
mlist = df[['MathScore', 'ReadingScore', 'WritingScore']]
print(mlist)
plt.title('Distribution of the test scores')
plt.show()
```

```python
ax = plt.subplots(figsize=(10, 6))
# This sets the number of bins in the histogram to 100, which controls how finely the data is divided for plotting.
# This enables the Kernel Density Estimate (KDE) plot, which overlays a smooth curve on top of the histogram to represent the distribution of the data more continuously.
sns.histplot(df[['MathScore', 'ReadingScore', 'WritingScore']], bins=100,  kde=True)
plt.title('Distribution of the test scores');
plt.show()
```
